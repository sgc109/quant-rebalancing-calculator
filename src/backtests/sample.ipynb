```kotlin
%use lets-plot
```

```kotlin
val rand = java.util.Random()
val data = mapOf(
    "rating" to List(200) { rand.nextGaussian() } + List(200) { rand.nextGaussian() * 1.5 + 1.5 },
    "cond" to List(200) { "A" } + List(200) { "B" }
)

var p = letsPlot(data)
p += geomDensity(color = "dark_green", alpha = .3) { x = "rating"; fill = "cond" }
p + ggsize(700, 350)
```

```kotlin
// Line
val start = 1
val stop = 100
val num = 100
val step = ((stop - start) / (num - 1))
val x = List(num) { start + it * step }
val lineData = mapOf(
    "x" to x,
    "y" to x.map { if (it % 2 == 0) 100 else -100 },
)

val line = letsPlot(lineData) { x = "x"; y = "y" } + geomLine() + ggtitle("Line")

line
```

```kotlin
val dateAndClosePriceList = csv.map {
    it["Date"] to it["Close"]
}.toMap()
dateAndClosePriceList
```

```kotlin
val lineData = mapOf(
    "date" to dateAndClosePriceList.keys.map { it.toString() },
    "price(USD)" to dateAndClosePriceList.values,
)

val line = letsPlot(lineData) { x = "date"; y = "price(USD)" } + geomLine() + ggtitle("Line")

line
```

```kotlin
val drawdownCalculator = DrawdownCalculator()
val y = drawdownCalculator.calculateDrawdowns(
    dateAndClosePriceList.values.map { it.toString().toDouble() },
).map { it * 100 }

val lineData = mapOf(
    "date" to dateAndClosePriceList.keys.drop(1).map { it.toString() },
    "drawdown(%)" to y,
)

val line = letsPlot(lineData) { x = "date"; y = "drawdown(%)" } + geomLine() + ggtitle("Line")

line
```
